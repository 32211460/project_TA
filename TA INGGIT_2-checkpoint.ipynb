```python
# Mengisi Missing Value Menggunakan Mean atau nilai rata_rata
def isi_data_hilang_dengan_rata_rata(df):
    for kolom in df.columns:
        if pd.api.types.is_numeric_dtype(df[kolom].dtype):  # Cek tipe data kolom
            rata_rata = df[kolom].mean()
            df[kolom].fillna(rata_rata, inplace=True)
            
# Memanggil fungsi untuk mengisi data yang hilang dengan rata-rata
isi_data_hilang_dengan_rata_rata(df)
            
# Tampilkan Data Missing Value yang sudah diisi dengan nilai rata-rata
print(df)
```

# Metode Elbow untuk Tingkat Kemiskinan di Pulau Jawa

```python
#mengambil nilai dari kolom 'Presentase penduduk miskin' (000)
X = df[['Presentase penduduk miskin']].values

#membuat sebuah list (daftar) dengan nama wcss
wcss = []

# menghitung nilai Within-Cluster Sum of Squares (WCSS) untuk berbagai jumlah cluster (dari 1 hingga 10).
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
    
#membuat sebuah grafik yang menampilkan hasil dari metode elbow (siku) dalam menentukan jumlah cluster yang optimal dalam menentukan jumlah cluster
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='-', color='b')
plt.title('Metode Elbow')
plt.xlabel('Jumlah Cluster')
plt.ylabel('WCSS (Within-cluster sum of squares)')
plt.grid(True)
plt.show()
```

```python
# Baca data
data = pd.read_excel('Data_Kemiskinan_Pulau_Jawa_cleaned.xlsx')

# fitur yang digunakan untuk clustering
features = ['Jumlah penduduk miskin (000)', 'Presentase penduduk miskin', 'Garis kemiskinan (Rp/Kap/Bulan)']

# Jumlah cluster
n_clusters = 3  

# Tetapkan nilai acak
np.random.seed(42)

# jumalah iterasi
max_iterations = 100

# memilih titik data secara acak sebagai pusat massa awal
initial_centroids_indices = np.random.choice(len(data), n_clusters, replace=False)
initial_centroids = data.loc[initial_centroids_indices, features].values

# melakukan iterasi pengelompokan
for iteration in range(max_iterations):
    # menghitung jarak dari setiap titik data ke setiap pusat massa
    distances = cdist(data[features], initial_centroids)
    
    # menemukan indeks centroid terdekat untuk setiap titik data
    nearest_centroid_indices = np.argmin(distances, axis=1)
    
    # memperbarui centroid berdasarkan cluster baru 
    centroids_new = data.groupby(nearest_centroid_indices)[features].mean().values

    # memeriksa Konvergensi
    if np.array_equal(initial_centroids, centroids_new):
        print(f"Converged after {iteration + 1} iterations.")
        break

    # memperbarui centroid untuk iterasi berikutnya
    initial_centroids = centroids_new

    # menambahkan jarak dan klaster yang ditetapkan ke DataFrame
    data[f'Jarak_C0_Iter{iteration + 1}'] = distances[:, 0]
    data[f'Jarak_C1_Iter{iteration + 1}'] = distances[:, 1]
    data[f'Jarak_C2_Iter{iteration + 1}'] = distances[:, 2]
    data[f'Cluster_Iter{iteration + 1}'] = nearest_centroid_indices

    # menghitung jarak terdekat untuk setiap titik data
    data[f'Jarak_Terdekat_Iter{iteration + 1}'] = np.min(distances, axis=1)

    # menampilkan DataFrame dengan jarak, cluster yang ditetapkan, dan jarak terdekat untuk setiap iterasi
    print(f"Data dengan Jarak ke Centroids (Iteration {iteration + 1}):")
    print(data[['Kabupaten/Kota', 'Jumlah penduduk miskin (000)', 'Presentase penduduk miskin',
                'Garis kemiskinan (Rp/Kap/Bulan)',
                f'Jarak_C0_Iter{iteration + 1}', f'Jarak_C1_Iter{iteration + 1}', f'Jarak_C2_Iter{iteration + 1}',
                f'Jarak_Terdekat_Iter{iteration + 1}', f'Cluster_Iter{iteration + 1}']])
    
    pd.set_option('display.max_rows', None)  # Menampilkan semua baris

print(df)
```
